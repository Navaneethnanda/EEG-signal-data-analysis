```
import numpy as np
```

```
import seaborn as sns
```

```
import matplotlib.pyplot as plt
```

```
from sklearn.metrics import classification_report, confusion_matrix
```

```
data=np.load("/content/drive/My Drive/Colab Notebooks/whole/data.npy", mmap_mode=None, allow_pickle=True, fix_imports=True, encoding='ASCII')
```

```
data.shape
```

```
labels=np.load("/content/drive/My Drive/Colab Notebooks/whole/labels.npy", mmap_mode=None, allow_pickle=True, fix_imports=True, encoding='ASCII')
```

```
labels.shape
```

# Valence

```
X=data
```

```
y1=labels[:,0]
```

```
y1=(y1>5).astype(int)
```

```
y1
```

```
from sklearn.preprocessing import StandardScaler
```

```
scalar=StandardScaler()
```

```
X=scalar.fit_transform(X)
```

```
from sklearn.model_selection import train_test_split
```

```
X_train,X_test,y1_train,y1_test=train_test_split(X,y1,test_size=0.2,random_state=22)
```

```
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=8)
model.fit(X_train, y1_train)
```

```
model.score(X_train,y1_train)
```

```
model.score(X_test,y1_test)
```

```
y_pred = model.predict(X_test)
```

```
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y1_test, y_pred))
print(classification_report(y1_test, y_pred))
```

```
error = []

for i in range(3, 20):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y1_train)
    pred_i = knn.predict(X_test)
    error.append(np.mean(pred_i != y1_test))
```

```
plt.figure(figsize=(12, 6))
plt.plot(range(3, 20), error, color='red', linestyle='dashed', marker='o',markerfacecolor='blue', markersize=10)
plt.title('Error Rate K Value')
plt.xlabel('K Value')
plt.ylabel('Mean Error')
```

```
optiKval=KNeighborsClassifier(n_neighbors=3)
optiKval.fit(X_train,y1_train)
```

```
ypred=optiKval.predict(X_test)
```

```
optiKval.score(X_test,y1_test)
```

```
print(confusion_matrix(y1_test, ypred))
print(classification_report(y1_test, ypred))
```
